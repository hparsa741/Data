### Introduction


A [Martingale](http://goo.gl/uk9vCE) is a betting strategy that the bettor doubles the wager if he loses so that the first win will recover all the losses plus a profit equal to the initial wager. A bettor with unlimited source of funds will eventualy win. This is the reasoning by the advocates of this strategy. However, there is always a limit to the weatlth of the bettor and the exponential growth of the wager will cause the bankruptcy of the bettor very soon. In the case of considering [house edge](http://en.wikipedia.org/wiki/Vigorish) where the game is not fair, the bankruptcy will happen even sooner. 
Another option is to change the wager multiplier and use a smaller or larger factor. In this Analysis we use Monte Carlo simulation to see the risk return trade off of such modification to the Matringale betting strategy.

`Julia` language is used for this analysis. The codes are available at [`GitHub`](https://github.com/hparsa741/Data/blob/master/Martingale_Monte_Carlo.jl). We start by loading the following package.

And the *multiplier* is used in the variable used for increasing/decreasing the wager in case of a loss in the previous gave.

### Variable wager

For a wager multiplier's range of 1 to 2.5 at steps of 0.05, we run each simulation for 10000 times. Each simulation is a sequence of games with *initial funds* of `10000`, *initial wager* of `100` and *wager count* of `200`. We consider the outcome of each sequence to be either gained profit denoted as *profitsCount* or bankcruptcy denoted as *bustsCount*. Then the average number of cases where the bettor gained profit or lost all of his funds is calculated as return and risk respectively. The case of finishing the wager count with terminal value less that initial funds has not been considered as risk in this analysis since the bettor has the chance of trying another round of wagering compared to bankruptcy where the bettor will be losing all his wealth.

```julia
global profArr = Float64[]
global bustArr = Float64[]
global multArr = Float64[]
srand(10)
for mult = 1:0.05:2.5 # trying a range of wager multiplier from 1 to 2.5 with step of 0.05
    bustsCount = 0
    profitsCount = 0
    sampleSize = 10000
    currentCount = 1
    while currentCount <= sampleSize
        Martingale(mult, 10000, 100, 200)
        currentCount += 1
    end
    push!(multArr, mult)
    push!(profArr, 100*profitsCount/sampleSize) # Average number of samples where bettor gained profit
    push!(bustArr, 100*bustsCount/sampleSize) # Average number of samples where bettor lost his funds
end
```

### Risk-return trade off

As a result of the above simulation, we have plot the profit versus bankruptcy rates. Knowing the profit and bankcruptcy rates, the bettor, as the decision maker can decide what are his excpetions on participating in this game. As depicted in the graph below, suppose the bettor is expecting to receive a profit rate of at least **`50`** percent. Also he is a risk avert at the rate of **`25`** percent. Consequently his selection is limited to the area between `magenta` vertical lines that is between wager multiplier `1.5` and `1.65`. This is his trade off area. For example if he decide to go with the maximum feasible return of 58, his risk would be maximum at the rate of 25. On the other hand, if he chose the minimum feasible risk of `12` percent, then the associate return of the wager will drop to `50` percent. The cyan vertical line here, shows the maximum return at wager multiplier `1.7` but the bankruptcy rate does not satisfy the bettor's risk criteria.

```julia
bustDf=DataFrame(A=[multArr], B=[bustArr], Metric="Broke")
profDf=DataFrame(A=[multArr], B=[profArr], Metric="Profit")
df= vcat(bustDf, profDf)

minProf = 50 # min Return set by bettor
maxBust = 25 # max Risk set by bettor

plot(layer(xintercept=[minimum(profDf[profDf[:B] .>= minProf, :A]), maximum(profDf[profDf[:B] .>= minProf, :A])], 
            Geom.vline(color="orange", size=.1mm)),
    layer(xintercept=[profDf[profDf[:B] .== maximum(profDf[:B]), :A]], Geom.vline(color="cyan", size=.1mm)),
    layer(yintercept=[minimum(profDf[profDf[:B] .>= minProf, :B]), maximum(bustDf[bustDf[:B] .<= maxBust, :B])], 
            Geom.hline(color="orange", size=.1mm)),
    layer(xintercept=[maximum(bustDf[bustDf[:B] .<= maxBust, :A])], Geom.vline(color="magenta", size=1mm)),
    layer(xintercept=[minimum(profDf[profDf[:B] .>= minProf, :A])], Geom.vline(color="magenta", size=1mm)),
    layer(df, x=df[:A], y=df[:B], color="Metric", Geom.line, Theme(line_width=.8mm)), 
Guide.xlabel("Multiplier"), 
Guide.ylabel("Rate %", orientation=:vertical),
Scale.discrete_color_manual("brown","darkgreen"))
```

### Wagering functions

```julia
# Wagering function of variable wager multiplier until wager count or all funds lost
function Martingale(multiplier, funds, initialWager, count)
    global bustsCount
    global profitsCount
    counter = 1
    wager = initialWager
    value = funds
    prevWager = "win"
    while counter <= count
        if prevWager == "win"
            wager = initialWager
            if rollDice()
                value += wager
                prevWager = "win"
            else
                value -= wager
                prevWager = "loss"
                if value <= 0
                    bustsCount += 1
                    break
                end
            end
        elseif prevWager == "loss"
            wager *= multiplier
            if (value - wager) < 0
                wager = value 
            end
            if rollDice()
                value += wager
                prevWager = "win"
            else
                value -= wager
                prevWager = "loss"
                if value <= 0
                    bustsCount += 1
                    break 
                end
            end
        end
        counter += 1
    end
    if value > funds
        profitsCount += 1
    end
end
```
